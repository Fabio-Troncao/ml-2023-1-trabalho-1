```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Lendo o arquivo
df = pd.read_excel('dataset.xlsx')

# Verificando dados faltantes, informo e substituindo por True
print(df.isnull().sum())
print(f'Quantidade de campos vazios ou nulos: {df.isnull().sum().sum()}')
df.fillna(True, inplace=True)

# Verificando inconsistências e dados inválidos
for column in df.columns:
    try:
        # Verificando o tipo de dados
        print(f'Tipo de dados para {column}: {df[column].dtype}')
        
        # Convertendo valores 'not_detected'/'detected' para numéricos
        dataf = df[column].replace({'not_detected': -1, 'detected': 1})
        
        # Verificando os valores mínimos e máximos
        print(f'Valor mínimo para {column}: {np.min(dataf)}')
        print(f'Valor máximo para {column}: {np.max(dataf)}')
        
        # Criando um histograma para visualizar os dados
        plt.hist(dataf, bins=50)
        plt.xlabel(column)
        plt.ylabel('Frequência')
        plt.show()
    except Exception as e:
        print(f'Erro na coluna {column}: {str(e)}')
```

```python
for column in df.columns:
    try:
        # Calculo os valores de Q1, Q3 e IQR para a coluna atual
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        # Calculo os limites inferior e superior para a detecção de outliers
        lim_inf = (1.5 * IQR) - Q1
        lim_sup = (1.5 * IQR) + Q3
        print(f'O limite inferior é: {lim_inf}')
        print(f'O limite superior é: {lim_sup}')

        # Identifico os outliers na coluna atual
        outliers = df[(df[column] < lim_inf) | (df[column] > lim_sup)]

        # Imprimo o número de outliers identificados na coluna atual
        print(f'Número de outliers na coluna {column}: {len(outliers)}')

        # Ploto um gráfico com os valores da coluna atual e os outliers destacados
        fig, ax = plt.subplots()
        ax.hist(df[column], bins=20)
        ax.set_title(column)

        if not outliers.empty:
            ax.plot(outliers[column], np.zeros_like(outliers[column]), 'ro', label='outliers')
            ax.legend()

        plt.show()

    except Exception as e:
        print(f'Erro na coluna {column}: {str(e)}')
```
